```python
#Function to enhance the user query

import requests
import time
import random

def enhance_query(query, api_key, max_retries=5):
    """
    Enhances a legal query for better retrieval accuracy using cosine similarity
    by expanding legal terms, synonyms, and related contexts.

    Implements exponential backoff for handling API failures.

    :param query: The original legal query (string)
    :param api_key: The Gemini API key (string)
    :param max_retries: Maximum number of retry attempts (default: 5)
    :return: The enhanced legal query (string)
    """
    url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash-latest:generateContent?key={api_key}"

    data = {
        "contents": [
            {
                "parts": [
                    {
                        "text": f"Rewrite the following legal query to maximize retrieval accuracy from the Indian Penal Code (IPC) using cosine similarity. "
                                f"Expand the query by incorporating relevant legal terms, synonyms, related offenses, and broader legal contexts. "
                                f"Ensure the language is precise and comprehensive to improve matching with relevant laws. "
                                f"Do not include specific article numbers or factual information—focus solely on enhancing the query's expressiveness. "
                                f"Increase the number of keywords so that cosine similarity retrieval is aided. "
                                f"Provide only the improved query, nothing else.\n\nQuery:\n{query}"
                    }
                ]
            }
        ]
    }

    headers = {"Content-Type": "application/json"}

    attempt = 0
    while attempt < max_retries:
        try:
            response = requests.post(url, headers=headers, json=data)

            if response.status_code == 200:
                response_data = response.json()
                improved_query = response_data.get("candidates", [{}])[0].get("content", {}).get("parts", [{}])[0].get("text", "").strip()
                return improved_query

            elif response.status_code in [500, 502, 503, 504]:  # Retryable server errors
                wait_time = 2 ** attempt + random.uniform(0, 1)  # Exponential backoff with jitter
                print(f"Server error {response.status_code}. Retrying in {wait_time:.2f} seconds...")
                time.sleep(wait_time)
                attempt += 1
            else:
                print(f"Error {response.status_code}: {response.text}")
                return None  # Non-retryable error

        except requests.exceptions.RequestException as e:
            wait_time = 2 ** attempt + random.uniform(0, 1)
            print(f"Request failed: {e}. Retrying in {wait_time:.2f} seconds...")
            time.sleep(wait_time)
            attempt += 1

    print("Max retries reached. Failed to enhance the query.")
    return None  # Return None if all retries fail

```

```python
import requests
from bs4 import BeautifulSoup

results_list = []  # Global list to store results

def scrape_articles(query):
    """Scrapes TOI articles related to the query and fetches content from LiveLaw if linked.
       Appends results to the global results_list."""

    def scrape_livelaw_article(url):
        """Scrapes an article from LiveLaw and returns its content."""
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers)

        if response.status_code != 200:
            return None

        soup = BeautifulSoup(response.text, 'html.parser')
        article_content = ""
        article_body = soup.find('div', class_='_s30J')  # Adjusted div class

        if article_body:
            for para in article_body.find_all(text=True, recursive=False):
                article_content += para.strip() + "\n"

        return article_content if article_content else None

    # Scrape TOI articles
    url = f"https://timesofindia.indiatimes.com/topic/{query}/news"

    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        return  # Exit if TOI request fails

    soup = BeautifulSoup(response.text, 'html.parser')
    first_div = soup.find_all('div', class_='crmK8')  # Adjust the class name if necessary

    for div in first_div:
        a_tags = div.find_all("a", href=True)

        for a in a_tags[:5]:  # Limit to top 5 articles
            href = a["href"]
            full_link = f"https://www.livelaw.in{href}" if href.startswith("/") else href

            content = scrape_livelaw_article(full_link)
            if content:
                results_list.append({"Article Heading": a.text.strip(), "Content": content})
    return results_list
```

```python
import requests
import json
import time

# Replace with your actual API key
api_key = gemini_api_key_2
url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash-latest:generateContent?key={api_key}"

def extract_relevant_points(query, results_list, max_retries=5):
    data = {
        "contents": [
            {
                "parts": [
                    {
                        "text": f"Given the following legal query, extract and arrange the most relevant points from the provided legal texts. "
                                f"Ensure the points are clear, structured, and directly address the query. "
                                f"Remove any redundant or unnecessary information.\n\n"
                                f"Query:\n{query}\n\n"
                                f"Legal Texts:\n{results_list}\n\n"
                                f"Provide the response in a structured, pointwise format."
                                f"List the results from IPC, BNS and the real cases seperately"
                                f"Make sure to add citations and also some real life cases along with citations relevant to it and given in the Legal text."
                                f"Make sure to include both BNS and IPC results if possible."
                                f"Also make sure to not use any of your own knowledge."
                    }
                ]
            }
        ]
    }

    headers = {"Content-Type": "application/json"}

    for attempt in range(max_retries):
        response = requests.post(url, headers=headers, json=data)

        if response.status_code == 200:
            response_data = response.json()
            return response_data.get("candidates", [{}])[0].get("content", {}).get("parts", [{}])[0].get("text", "").strip()

        elif response.status_code in {429, 500, 502, 503, 504}:  # Retry-worthy errors
            wait_time = 2 ** attempt  # Exponential backoff
            print(f"Attempt {attempt + 1}: Error {response.status_code}. Retrying in {wait_time}s...")
            time.sleep(wait_time)
        else:
            return f"Error: {response.status_code}\n{response.text}"

    return "Failed after multiple retries."
```

```python
# --- Main Conversational Loop ---

def legal_chatbot():
    """Runs the legal chatbot."""
    print("Welcome to the Legal Chatbot!  Ask me about Indian Law.")

    conversation_history = []  # Store conversation turns
    api_key_choice = gemini_api_key_1  # Default API key

    while True:
        user_query = input("You: ")
        if user_query.lower() == "exit":
            print("Goodbye!")
            break

        # Enhance Query
        enhanced_query = enhance_query(user_query, api_key_choice, max_retries=3)

        if not enhanced_query:
            print("Chatbot: Sorry, I couldn't process your query. Please try again.")
            continue

        # Embed Query
        embedding = embed_query(enhanced_query)

        # Query Pinecone
        pinecone_results = query_pinecone(embedding, top_k=5)  # Reduced top_k for faster response

        # Scrape Articles
        article_results = scrape_articles(user_query)
        # Handle NoneType returns from Pinecone and scraping

        if pinecone_results is None:
            pinecone_results = []

        if article_results is None:
            article_results = []

        combined_results = pinecone_results + [str(article) for article in article_results]

        # Extract Relevant Points
        response_text = extract_relevant_points(user_query, combined_results, max_retries=3)

        #Basic error handling.
        if "Error" in response_text or "Failed after multiple retries" in response_text:
            print("Chatbot: I encountered an issue while processing your request. Please try again with a different query.")
            continue

        print("Chatbot:", response_text)
        print()

        #Conversation history is updated.
        conversation_history.append({"user": user_query, "bot": response_text})

        #Post-response questioning loop:
        while True:
          follow_up = input("\n\nDo you have a follow-up question about this? (yes/no): ").lower()
          if follow_up == "no":
            break #Exit follow-up loop.
          elif follow_up == "yes":
            follow_up_query = input("\n\nYour follow-up question: ")

            # Enhance Follow-Up Query - Use conversation history for context
            contextual_query = f"Previous query: {user_query}. Previous answer: {response_text}. Follow-up question: {follow_up_query}"
            enhanced_follow_up = enhance_query(contextual_query, api_key_choice, max_retries=3)

            if not enhanced_follow_up:
              print("Chatbot: Sorry, I couldn't process your follow-up query. Please try again.")
              continue

            # Embed Follow-Up Query
            follow_up_embedding = embed_query(enhanced_follow_up)

            # Query Pinecone with Follow-Up
            follow_up_pinecone_results = query_pinecone(follow_up_embedding, top_k=5)

            # Scrape Articles (consider skipping if very recent)
            follow_up_article_results = scrape_articles(follow_up_query)
            # Handle NoneType returns from Pinecone and scraping

            if follow_up_pinecone_results is None:
                pinecone_results = []

            if follow_up_article_results is None:
                article_results = []

            combined_follow_up_results = pinecone_results + [str(article) for article in article_results]
            # combined_follow_up_results = follow_up_pinecone_results + [str(article) for article in follow_up_article_results]

            # Extract Relevant Points for Follow-Up
            follow_up_response = extract_relevant_points(follow_up_query, combined_follow_up_results, max_retries=3)

            if "Error" in follow_up_response or "Failed after multiple retries" in follow_up_response:
              print("Chatbot: I encountered an issue while processing your follow-up request. Please try again with a different question.")
              continue

            print("Chatbot:", follow_up_response)
            print()
            conversation_history.append({"user": follow_up_query, "bot": follow_up_response}) #add to history


          else:
            print("Please answer 'yes' or 'no'.")

# --- Run the Chatbot ---
# legal_chatbot()
```

```python
import gradio as gr
import traceback

# --- Chatbot with Gradio Interface ---
def legal_chatbot_gradio(query, history):
    """Processes the legal query and generates a response with conversation history."""
    try:
        # Enhance Query
        enhanced_query = enhance_query(query, gemini_api_key_1, max_retries=3)

        if not enhanced_query:
            return "Chatbot: Sorry, I couldn't process your query. Please try again."

        # Embed Query
        embedding = embed_query(enhanced_query)

        # Query Pinecone
        pinecone_results = query_pinecone(embedding, top_k=5)

        # Scrape Articles
        article_results = scrape_articles(query)

        # Handle NoneType returns
        if pinecone_results is None:
            pinecone_results = []

        if article_results is None:
            article_results = []

        combined_results = pinecone_results + [str(article) for article in article_results]

        # Extract Relevant Points
        response_text = extract_relevant_points(query, combined_results, max_retries=3)

        # Basic error handling
        if "Error" in response_text or "Failed after multiple retries" in response_text:
            return "Chatbot: I encountered an issue while processing your request. Please try again with a different query."

        # Update history
        history.append((query, response_text))  # Append user query and response to history
        return response_text

    except Exception as e:
        error_message = f"Chatbot: An unexpected error occurred.\n{traceback.format_exc()}"
        print(error_message)  # Print the full traceback to the console for debugging
        return "Chatbot: Sorry, an internal error occurred.  Check the console for details."



# Gradio Interface
iface = gr.ChatInterface(
    fn=legal_chatbot_gradio,
    title="Indian Legal Chatbot",
    description="Ask questions about Indian law.",
    examples=["What are the consequences of theft?", "Explain defamation laws in India.", "What is cyberbullying and its punishment?"],
    chatbot=gr.Chatbot(height=500),  # Add height parameter
)

iface.launch()
```
